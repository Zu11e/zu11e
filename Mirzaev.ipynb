```python
# Загружаем нужные библиотеки

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
flights = pd.read_csv(r"C:/Users/xewte/Desktop/Биг Дата/flights.csv")
airports = pd.read_csv(r"C:/Users/xewte/Desktop/Биг Дата/airports.csv")
airlines = pd.read_csv(r"C:/Users/xewte/Desktop/Биг Дата/airlines.csv")
```

```python
flights.columns = flights.columns.str.lower()
```

```python
flights
```

```python
number_flights_per_month = flights.day.groupby(flights.month).count()
number_flights_per_month = number_flights_per_month.reset_index()
number_flights_per_month.rename(columns={'day': 'flights'}, inplace=True)
number_flights_per_month
```

```python
x = number_flights_per_month.month
y = number_flights_per_month.flights

fig, ax = plt.subplots()

ax.bar(x, y)

ax.set_xlabel('Месяцы',fontsize = 15)
ax.set_ylabel('Рейсы',fontsize = 15)

fig.set_figwidth(20)    #  ширина Figure
fig.set_figheight(10)    #  высота Figure

plt.show()
```

Построили график кол-ва рейсов по месяцам.


```python
flights_day_of_the_week = flights.day.groupby(flights.day_of_week).count()
flights_day_of_the_week = flights_day_of_the_week.reset_index()
flights_day_of_the_week.rename(columns={'day': 'flights'}, inplace=True)
flights_day_of_the_week
```

```python
x = flights_day_of_the_week.day_of_week
y = flights_day_of_the_week.flights

fig, ax = plt.subplots()

ax.bar(x, y)

ax.set_xlabel('Дни недели',fontsize = 15)
ax.set_ylabel('Рейсы',fontsize = 15)

fig.set_figwidth(20)    #  ширина Figure
fig.set_figheight(10)    #  высота Figure

plt.show()
```

```python
number_flights_per_year = flights.day.groupby(flights.origin_airport).count()
#
#number_flights_per_year.rename(columns={'day': 'flights'}, inplace=True)
number_flights_per_year
```

```python
destination_per_year = flights.day.groupby(flights.destination_airport).count()
#destination_per_year.rename(columns={'day': 'flights'}, inplace=True)
destination_per_year=destination_per_year.append(number_flights_per_year)
destination_per_year=pd.DataFrame(destination_per_year)
destination_per_year = destination_per_year.day.groupby(destination_per_year.index).sum()
destination_per_year
```

```python
top_10 = airports.set_index("IATA_CODE").join(destination_per_year)
top_10.day.groupby(top_10.CITY).sum().nlargest(10)
```
